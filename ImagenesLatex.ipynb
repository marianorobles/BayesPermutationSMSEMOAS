```python
import os
import glob
import re
```

```python
def descripcion(nombre):
    partes=re.split(pattern='_',string=nombre)
    desc=''
    if len(partes)>1:
        probl=partes[-4]
        ind=partes[-3]
        dim=re.search(pattern='\d',string=partes[-2])[0]
        tipo=re.search('[a-z]+',partes[-1])[0]
        if tipo=='to':
            desc='Probabilidad de ser el mejor algoritmo '+probl+' con indicador '+ind+' con '+dim+' objetivos.'
            return desc,'b'#estadísticos bayesianos
        if tipo=='gc':
            tipo='Grafico de convergencia'
        elif tipo=='fp':
            tipo='Aproximación al frente de pareto'
        elif tipo=='bp':
            tipo='Diagrama de caja'
        alg=''
        if len(partes)==5:
            alg=' usando algoritmo'+partes[0]
        desc=tipo+' del problema '+probl+' con indicador '+ind+alg+' con '+dim+' objetivos.'
        return desc,'e'#estadísticos descriptivos
    return desc,''
```

```python
#nombres de los archivos
archivos=[''.join(os.path.splitext(os.path.basename(dire))) for dire in glob.glob("./img/*")]
```

```python
contador=0
est_desc=[]
est_bay=[]
for archivo in archivos:
    res=descripcion(archivo)
    if res[1]=='b':
        est_bay.append(res[0])
    elif res[1]=='e':
        est_desc.append(res[0])
for texto in est_desc:
    if contador%3==0:
        print('\clearpage')
    print('¿?begin{figure}[0.5]')
    print('¿?includegraphics[width=0.5¿?textwidth]{'+archivo+'}')
    print('¿?caption{'+texto+'}')
    print('¿?end{figure}')
    contador+=1
for texto in est_bay:
    if contador%3==0:
        print('\clearpage')
    print('¿?begin{figure}[0.5]')
    print('¿?includegraphics[width=0.5¿?textwidth]{'+archivo+'}')
    print('¿?caption{'+texto+'}')
    print('¿?end{figure}')
    contador+=1
```
