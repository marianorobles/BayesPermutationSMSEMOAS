```python
#importamos algoritmos
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.algorithms.moo.nsga3 import NSGA3
from pymoo.algorithms.moo.moead import MOEAD
from pymoo.algorithms.moo.sms import SMSEMOA
#importamos indicadores
from pymoo.indicators.igd_plus import IGDPlus
from pymoo.indicators.igd import IGD
from pymoo.indicators.gd_plus import GDPlus
from pymoo.indicators.gd import GD
#importamos función para resolver problemas de optimización
from pymoo.optimize import minimize
#importamos funciones para visualizar resultados
from pymoo.util.display.column import Column
from pymoo.util.display.output import Output
#usamos pandas para nuestra base de datos
import pandas as pd
#numpy para algunos cálclulos
import numpy as np
#importamos función para traer problemas
from pymoo.problems import get_problem
#función para ref_dirs
from pymoo.util.ref_dirs import get_reference_directions
#función para detener algortimo
from pymoo.termination import get_termination
#time para generar la primer semilla
import time
```

```python
class Salida(Output):
    def __init__(self):
        super().__init__()
        #definimos columnas
        self.pf='a'
        self.gdp=Column("gd+",width=13)
        self.f_igdp = Column("igd+", width=13)
        self.delta_p=Column("delta_p",width=13)
        self.time = Column("time", width=13)
        self.x=Column("X",width=13)
        self.f=Column("F",width=13)
        
        #añadimos las columnas a las ya existentes
        self.columns += [self.gdp,self.delta_p,self.f_igdp,self.time,self.x,self.f]
        

    def update(self, algorithm):
        
        super().update(algorithm)
        if self.pf=='a':
            self.pf=pd.read_csv('./DataWfg/'+algorithm.problem.name()+'_24_10.txt',header=None).values
        #usamos igdp con el frente de pareto
        F=algorithm.opt.get('F')
        #self.f_igd_p.set(ind(algorithm.opt.get("F")))
        ind = IGD(self.pf)
        igd_p=((1/len(self.pf)**(1/2)))*len(self.pf)*ind(F)
        ind = GD(self.pf)
        gd_p=(1/len(F)**(1/2))*len(F)*ind(F)
        self.delta_p.set(max(igd_p,gd_p))
        #self.delta_p.set(max(igd_p,gd_p))
        ind=GDPlus(self.pf)
        self.gdp.set(ind(algorithm.opt.get("F")))
        ind = IGDPlus(self.pf)
        self.f_igdp.set(ind(algorithm.opt.get("F")))
        #tiempo de ejecución
        self.time.set(algorithm.result().exec_time)
        #resultados
        self.f.set(F)
        self.x.set(algorithm.opt.get('X'))
#gd cardinalidad de aproximacion
#igd cardinalidad de frente
```

```python
problemas=["wfg1","wfg2","wfg3","wfg4","wfg5","wfg6","wfg7","wfg8","wfg9"]
num_total=20*len(problemas)
semillas=np.random.randint(1,10000,num_total+1)
```

```python
ref_dirs=get_reference_directions("energy", objetivos, 200, seed=semillas[-1])#cambiar segundo parametro
algoritmos={'NSGA2':NSGA2(pop_size=200,prob_neighbor_mating=0.8,n_neighbors=10),
            'NSGA3':NSGA3(pop_size=200,ref_dirs=ref_dirs,prob_neighbor_mating=0.8,n_neighbors=10),
            'MOEAD':MOEAD(ref_dirs=ref_dirs,prob_neighbor_mating=0.8,n_neighbors=10)}
termination = get_termination("n_gen", 500)
```

```python
lista=[]
j=0
for problema in problemas:
    for algoritmo in algoritmos:
        #k se reinicia en cada algoritmo nuevo para que tengan la misma semilla diferentes algoritmos
        k=0
        for i in range(20):
            semilla=semillas[j+k]
            lista.append([algoritmo,problema,i,semilla,j])
            k+=1
    j+=k
```

```python
comb=dict(zip(range(len(lista)),lista))
comb
```

```python
inicio=537
for num in range(inicio,len(comb)-1):#desde el punto necesario hasta el último
    print(comb[num])
    algoritmo=comb[num][0]
    problema=comb[num][1]
    semilla=comb[num][3]
    i=comb[num][2]
    j=comb[num][4]
    algorithm = algoritmos[algoritmo]
    problem=get_problem(problema,n_var=24,n_obj=objetivos)
    result=minimize(
        problem=problem,algorithm=algorithm,termination=termination,
        verbose=True,output=Salida(),save_history=True,seed=semilla
    )
    lista=[[columna.value for columna in observacion.output.columns] for observacion in result.history ]
    temporal=pd.DataFrame(lista,columns=["Epoca","Fe","gdp","delta_p","igdp","Tiempo","X","F"])
    temporal['Ejecucion']=i+1
    temporal['Algoritmo']=algoritmo
    temporal['Problema']=problema
    temporal.to_csv('resultados'+str(objetivos)+'_dim.csv',index=False,mode='a',header=False,
                    columns=["Epoca","Fe","gdp","delta_p","igdp","Tiempo","X","F","Ejecucion","Algoritmo","Problema"])
    x=pd.DataFrame(result.algorithm.opt.get('X'))
    x.to_csv('x'+str(objetivos)+'_dim.csv',index=False,mode='a')
    metadatos=pd.DataFrame([[algoritmo,problema,i+1]])
    metadatos.to_csv('metadatos'+str(objetivos)+'_dim.csv',index=False,mode='a')
```
