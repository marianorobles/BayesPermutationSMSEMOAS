```python
from pymoo.util.display.column import Column
from pymoo.util.display.output import Output
class MyOutput(Output):

    def __init__(self):
        super().__init__()
        #self.pf = get_problem("dtlz2").pareto_front()
        self.f_igdp = Column("igd+", width=13)
        self.time=Column("time",width=13)
        self.f=Column("F(X)",width=13)
        self.x=Column("X",width=13)
        self.columns += [self.f_igdp,self.time,self.f,self.x]

    def update(self, algorithm):
        super().update(algorithm)
        #print("IGD+", ind(A))
        pf=algorithm.problem.pareto_front()
        ind = IGDPlus(pf)
        self.f_igdp.set(ind(algorithm.opt.get("F")))
        self.time.set(algorithm.result().exec_time)
        self.f.set(algorithm.opt.get("X"))
        self.x.set(res.algorithm.opt.get('F'))
        #self.time=algorithm.end_time
        #self.x_hv.set(ind(algorithm.opt.get("F")))
```
